```python
import cv2
import os
import numpy as np
import time
```

```python
path = "" #define your path to store the images
```

```python
#function to Display image
def Dis(img):
    cv2.imshow('img',img)
    k =cv2.waitKey(10) # if you want more time increase the waitkey value complexity will also be increased by that much factor
    return 1
```

```python
def makedir(path):
    if os.path.isdir(path) and os.listdir(path)!=0  :
        print("enter another name your name is registerd ")
    else :
        if os.path.exists(path):
            if os.listdir(path)==0:
                os.rmdir(path)

        os.mkdir(path)
```

```python
#this function gives you the time to change hands while capturing the images and swapping the hands
def test(cam,k,hand):
    c=0
    while cam.isOpened():
        x,frame = cam.read()
        #print(x,frame)
        if x :
            #frame   = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)  # if you want Binary image uncomment this line
            frame   = cv2.resize(frame,(900,500),interpolation=cv2.INTER_LINEAR)
            frame = cv2.flip(frame,1)#removes mirror effect of camera
            if hand=='r':
                cv2.rectangle(frame,(600,0),(900,300),(255,0,0),4) #choose blue window size as per your requirment
            else :
                cv2.rectangle(frame,(0,0),(300,300),(255,0,0),4)
            cv2.imshow('img',frame)
            cv2.waitKey(100)
            c = c+1
            #print(c)
            if c==k:
                break

    cv2.destroyAllWindows()
```

## Main Code Starts Here

```python
# Actual Main Function
cam = cv2.VideoCapture(0,cv2.CAP_DSHOW)

while True:
    hand = 'r'
    dname = input("Enter dir name ")
    makedir(path+dname)
    k = int(input("No of Samples\n"))
    c = 0
    test(cam,50,'r')
    time.sleep(2)
    while cam.isOpened():
        x,frame = cam.read()
        #print(x,frame)
        if x :
            #frame   = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            frame   = cv2.resize(frame,(900,500),interpolation=cv2.INTER_LINEAR)
            frame = cv2.flip(frame,1)#removes mirror effect of camera
            if hand=='r':
                cv2.rectangle(frame,(600,0),(900,300),(255,0,0),4)
            else :
                cv2.rectangle(frame,(0,0),(300,300),(255,0,0),4)
            if(Dis(frame)==1):
                pass
            else :
                Dis(frame)
            if hand=='r':
                cv2.imwrite(path+dname+"/img "+str(c)+'.jpg',frame[0:300,600:900,:])
            else :
                cv2.imwrite(path+dname+"/img "+str(c)+'.jpg',frame[0:300,0:300])
            c = c+1
            #print(c)
            if c==k/2:
                hand = input("change hand if needed")
                time.sleep(2)
                test(cam,50,hand)
            elif c==k:
                break
    cv2.destroyAllWindows()
    b = int(input("ente 1 to exit "))
    if b== 1:
        break
cam.release()
```
